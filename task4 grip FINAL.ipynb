## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading Dataset

```python
df=pd.read_csv("Downloads/iris.csv")
df.head()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Species"] = le.fit_transform(df["Species"])
df.head()
```

```python
x=df.iloc[:,[1,2,3,4]]
y = df.iloc[:,5]
```

## Visualizing Dataset

```python
x1=df["SepalLengthCm"]
y1=df["Species"]
with sns.axes_style('white'):
    sns.jointplot(x1, y1, df, kind='kde')
```

```python
x1=df["SepalWidthCm"]
y1=df["Species"]
sns.swarmplot(y1, x1)
```

```python
sns.barplot(x="PetalLengthCm",y="Species",data=df,estimator=np.mean)
```

```python
sns.catplot(x="PetalWidthCm",y="Species",data=df)
```

## Exploring Decision Tree Algorithm

```python
from sklearn import tree
m= tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
(xtrain,xtext,ytrain,ytest)= train_test_split(x,y,test_size = 0.2, random_state =0)
```

```python
m.fit(xtrain,ytrain)
m.score(xtrain,ytrain)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dtree= DecisionTreeClassifier().fit(x, y)
plt.figure(figsize = (20,10))
plot_tree(dtree, rounded = True, filled=True)
plt.show()
```
